```python
import geopandas as gpd
import pandas as pd
```

```python
d1=pd.read_csv("E:/sem5/thesis/results/multi_18_01.csv")#....first set of gee table
```

```python
len(d1["point_id"].unique())
```

```python
d2=pd.read_csv("E:/sem5/thesis/results/lucas_date.csv")#....ground truth table
```

```python
d2
```

```python
d2["date_format"]=pd.to_datetime(d2["date_format"])
```

```python
d1["date_format"]=pd.to_datetime(d1["date_format"])
```

```python
#selecting row with same/most recent date with groud truth points.
from datetime import date, timedelta
less=timedelta(days=365)
df=pd.DataFrame()
for j in range(len(d2['point_id'])):
    tem_df=(d1.loc[(d1['point_id'] == d2['point_id'][j])]).reset_index()
    for i in range(len(tem_df['point_id'])):
        diff=abs((d2["date_format"][j])-(tem_df["date_format"][i]))
        if(diff<less):
            less=diff
            #print(less) 
            df1=tem_df.loc[(tem_df['date_format'] == tem_df['date_format'][i])]
    
    if len(tem_df)!=0:
        df=pd.concat([df,df1], ignore_index=True)
```

```python
df.to_csv("E:/sem5/thesis/results/lucas_1000.csv")
```
